```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv")
df.head()
```

```python
x = df['T2']
y = df['T6']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train.values.reshape(-1,1), y_train.values.reshape(-1,1))
prediction = lin_reg.predict(x_test.values.reshape(-1,1))
```

```python
#number12
from sklearn.metrics import r2_score
r2_score = r2_score(prediction, y_test)
r2_score
```

```python
#number13
x = df.drop(['date', 'lights', 'Appliances'], axis=1)
y = df['Appliances']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scale = scaler.fit_transform(x)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scale, y, test_size=0.3, random_state=42)

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
prediction = lin_reg.predict(x_test)

from sklearn.metrics import mean_absolute_error
mean_absolute = mean_absolute_error(prediction, y_test)
round(mean_absolute,2)
```

```python
#number 14 residual sum
df_residual = pd.DataFrame({'Actual': y_test, 'Predicted':prediction})
residual = np.sum(np.square(df_residual['Predicted'] - df_residual['Actual']))
print(' residual sum of squares is : ', round(residual,2)) 
```

```python
#number15 Root mean square
from sklearn.metrics import mean_squared_error
RMSE = np.sqrt(mean_squared_error(prediction, y_test))
round(RMSE,2)
```

```python
#number16
from sklearn.metrics import r2_score
r2_score = r2_score(prediction, y_test)
round(r2_score,2)
```

```python
#number17
def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', 'col_name']
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = get_weights_df(lin_reg, x,x)
linear_model_weights
```

```python
#number19 Train lasso regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)

lasso_weights_df = get_weights_df(lasso_reg, x,x)
lasso_weights_df
```

```python
#number20
from sklearn.metrics import mean_squared_error
lasso_pred = lasso_reg.predict(x_test)
RMSE_lasso = np.sqrt(mean_squared_error(lasso_pred, y_test))
round(RMSE_lasso,3)
```
